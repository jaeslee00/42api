```python
import json
from time import sleep
from ftapi import *

u,s = auth()
jai = Api(u,s)

file = open("piscine.txt")
data = json.loads(file.read())
```

```python
idx = 1
user_data = []
while True:
    down = jai.campus_users('1', filter={"pool_year":"2019", "pool_month":"july"}, page={"size":100, "number":idx}).get()
    if len(down) < 1:
        break
    user_data = user_data + down
    idx += 1
    sleep(0.5)
```

```python
user_list = {}
for idx, elem in enumerate(user_data):
    if elem["login"].find("3b3-") > -1:
        pass
    else:
        user_list.update({elem["login"]: elem["id"]})
for user_login in user_list.keys():
    print(user_login)
```

```python
score_data = {}
for idx, user_login in enumerate(user_list.keys()):
    resp = jai.users(user_login).get()
    sleep(0.5)
    scores = {}
    for p in resp["projects_users"]:
        if p["project"]["slug"].find("c-piscine") > -1:
            scores.update({p["project"]["slug"]: p["final_mark"]})
    score_data.update({user_login: scores})
    print(idx)
    if idx > 30:
        break
```

```python
correction_data = {}
for user in user_data:
    correctors_data.update({user["login"]:[]})
```

```python
for user in user_data:
    for project in peer_projects:
        team = jai.users_teams(str(user['id']), filter={"project_id":project["id"]}).get()
        sleep(0.5)
        for scale_team in team:
            for correction in scale_team["scale_teams"]:
                if correction["corrector"] == 'invisible' or not scale_team["teams_uploads"] or correction["final_mark"] is None:
                    continue
                corrector = correction["corrector"]["login"]
                given_mark = correction["final_mark"]
                moulinette_mark = scale_team["teams_uploads"][0]["final_mark"]
                diff = abs(given_mark - moulinette_mark)
                if corrector in correctors_data:
                    correctors_data[corrector].append(
                            {
                                "project_id": project["id"],
                                "project_name": project["slug"],
                                "diff": diff
                            })
```
